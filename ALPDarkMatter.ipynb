```python
fig,ax = FigSetup(Shape='Square',ylab=r'$1/f_a$ [GeV$^{-1}$]',mathpazo=True,\
                              g_min=1e-14,g_max=1e-11,m_min=1e-12,m_max=1e-3,xtick_rotation=0)

BlackHoleSpins(ax,1,[0.6e-13,0.4e-17],text_on=False,alpha=1,facecolor='gray')
Axion_fa.QCDAxion(ax,QCD_label_mass=3e-6,shading_on=False,text_col='k',text_on=False)
Axion_fa.SN1987A(ax,text_pos=[4e-5,0.35e-8])

def fa_TD_ALP(m,n,theta_eff):
    m_pl = 2.4e18
    gstar = 75
    cn = (2/5)*(1+n/5)
    npow = (n/(2*(n+4)))
    sqrt_m0m1 = (m_pl/(cn*1.66*sqrt(gstar)))**npow
    fa = (1e12**2*1.265/(9e-3*theta_eff**2*sqrt(m/1e-6)*sqrt_m0m1))**(1/(2-npow))
    return fa

def fa_TD_ALP_alt(m,n,theta_eff,tau_star=16):
    m_pl = 2.4e18
    gstar = 75
    sqrt_m0m1 = tau_star**(n/4)
    fa = (1e12**2*1.265/(9e-3*theta_eff**2*sqrt(m/1e-6)*sqrt_m0m1))**(1/2)
    return fa

th_eff = [2.33,2.02,1.81,1.69,1.59,1.56,1.48]

cols = cmr.pride(linspace(0.15,0.85,7))

m = array([1e-12,1e2])

zo = -100
for i in range(0,7):
    plt.plot(m,fa_TD_ALP(m,i,th_eff[i])**-1,'-',lw=3,path_effects=line_background(5,'k'),color=cols[i,:],zorder=-50)
    plt.fill_between(m,fa_TD_ALP(m,i,th_eff[i])**-1,y2=fa_TD_ALP(m,6,th_eff[6])**-1,color=cols[i,:],zorder=zo)
    zo += 1


    
    

dat = loadtxt('limit_data/AxionPhoton/Projections/DMRadio.txt')
plt.fill_between(dat[:,0],dat[:,1]*(137*2*pi),y2=1e0,zorder=0,lw=3,color='crimson',alpha=0.1)
plt.plot(dat[:,0],dat[:,1]*(137*2*pi),'k:',zorder=0,lw=3,color='crimson')
plt.text(3e-7,2e-14,r'{\bf DM-Radio}',color='crimson',rotation=72,fontsize=22)
plt.text(6e-7,2.1e-14,r'$C_{a\gamma}=1$',color='crimson',rotation=72,fontsize=22)

dat = loadtxt('limit_data/AxionNeutron/Projections/CASPEr_wind.txt')
dat[:,1] *= 2*AxionNeutron.m_n
plt.fill_between(dat[:,0],dat[:,1]/AxionNeutron.m_n,y2=1e0,zorder=-10,lw=3,color='darkred',alpha=0.1)
plt.plot(dat[:,0],dat[:,1]/AxionNeutron.m_n,'k--',zorder=-10,lw=3,color='darkred')
plt.text(2.5e-9,8e-13,r'{\bf CASPEr-gradient}',color='darkred',rotation=59,fontsize=22)
plt.text(0.9e-8,1e-12,r'$C_{an}=1$',color='darkred',rotation=59,fontsize=22)


# https://arxiv.org/abs/hep-ph/9908508
Axion_fa.CASPEr(ax,text_pos=[0.4e-8,1.2e-14],rot=80,col='#a83279')
plt.text(0.8e-8,1.2e-14,r'$C_{\rm EDM} = 2.4\times 10^{-16}$ cm',color='#a83279',rotation=80,fontsize=18)
    
    
plt.text(6e-5,3.3e-12,r'$n = 6$',color=cols[6,:],rotation=40,fontsize=35)
plt.text(9e-5,0.48e-12,r'$n = 0$',color=cols[0,:],rotation=36,fontsize=35)
plt.text(4e-6,2e-12,r'{\bf QCD axion}',fontsize=35,rotation=70)
plt.text(2e-12,0.3e-13,r'{\bf Black hole spins}',fontsize=35,color='k',rotation=90)


plt.gcf().text(0.88,0.15,r'$\frac{m_a}{m_a(T_1)} = \frac{T_1}{\sqrt{m_a f_a}}$',ha='right',fontsize=30)


MySaveFig(fig,'fa_ALPDM')
```
