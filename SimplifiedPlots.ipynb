```python
fig,ax = FigSetup(Shape='Square',ylab=r'Axion-electron coupling',xlab='Axion mass [eV]',mathpazo=True,\
                  g_min=1e-15,g_max=1e-10,m_min=1e-6,m_max=1e5,xtick_rotation=0)

#AxionElectron.QCDAxion(ax,Hadronic_on=False,KSVZ_on=False,text_on=False)


g_min=1e-15
g_max=1e-10
m_min=1e-6
m_max=1e5

# Mass-coupling relation
def g_x(C_ae,m_a):
    return 8.943e-11*C_ae*m_a
cmap = 'YlOrBr'
C_center = 10.0**((log10(2.0e-5)+log10(1.0/3.0))/2)
C_width = 2
C_logwidth = 0.3
vmax = 0.9

# Plot Band
n = 200
g = logspace(log10(g_min),log10(g_max),n)
m = logspace(log10(m_min),log10(m_max),n)
QCD = zeros(shape=(n,n))
for i in range(0,n):
    QCD[:,i] = norm.pdf(log10(g)-log10(g_x(C_center,m[i])),0.0,C_width)
cols = cm.get_cmap(cmap)

cols.set_under('w') # Set lowest color to white
vmax = amax(QCD)*1.5
vmin = 0.12
plt.contourf(m, g, QCD, 50,cmap=cols,vmin=vmin,vmax=vmax,zorder=0)
plt.contourf(m, g, QCD, 50,cmap=cols,vmin=vmin,vmax=vmax,zorder=0)
plt.contourf(m, g, QCD, 50,cmap=cols,vmin=vmin,vmax=vmax,zorder=0)


plt.text(0.9e-2,0.4e-14,r'{\bf QCD axion}',fontsize=40,color='goldenrod',rotation=63,alpha=0.9,path_effects=line_background(1.5,'k'))

MySaveFig(fig,'AxionElectron_Simple_QCDAxionOnly')
```
