# Axion coupling to neutron electric dipole moment

QCD axion line defined by
\begin{equation}
m_{a}=5.70(7) \mu \mathrm{eV}\left(\frac{10^{12} \,\mathrm{GeV}}{f_{a}}\right)
\end{equation}

This can be related to the EDM coupling (the characteristic coupling of the QCD axion) via:
\begin{equation}
g_{a n \gamma}=-g_{a p \gamma}=e \frac{C_{\mathrm{EDM}}}{f_{a}}=(3.7 \pm 1.5) \times 10^{-3}\left(\frac{1}{f_{a}}\right) \frac{1}{\mathrm{GeV}}
\end{equation}
The shading the QCD line reflects this theoretical uncertainty, see https://arxiv.org/abs/1306.6088

It must be noted that this plot and the 1/fa plot are, at best, challenging to interpret beyond a purely phenomenological point of view, and are at worst highly misleading. Most crucially, the area outside of the QCD line requires significant (and not always natural!) model-building. In essence one needs a "QCD" axion whose mass is different from than the mass arising just from the anomalous gluon coupling of the axion (which defines 1/f_a). There are very few models that achieve this naturally, such as the Z_N axion model of Hook (https://arxiv.org/abs/1802.10093), and developed phenomenologically by Di Luzio et al. (https://arxiv.org/pdf/2102.00012.pdf)
