```python
# Figure setup
fig,ax = FigSetup(Shape='Rectangular',xlab='Scalar mass, $m_\phi$ [eV]',ylab=r'Scalar-EM coupling, $d_e$',mathpazo=True,\
                              g_min=1e-13,g_max=1e10,m_min=1e-24,m_max=1e1)

ScalarPhoton.EotWashEP(ax)
ScalarPhoton.MICROSCOPE(ax)


dat = flipud(loadtxt('limit_data/ScalarBaryon/FifthForce_CI.txt'))
m1 = 0.1973*1e-6/dat[:,0]
d_e1 = 500*sqrt(dat[:,1])

dat = flipud(loadtxt('limit_data/ScalarBaryon/FifthForce_Mars.txt'))
m2 = 0.1973*1e-6/dat[:,0]
d_e2 = 500*sqrt(dat[:,1])

dat = loadtxt('limit_data/ScalarNucleon/Union_InverseSquareLaw.txt')
m3 = dat[:,0]
d_e3 = 500*dat[:,1]*sqrt(1.37e37)

mvals = logspace(-24,2,1000)
dvals1 = 10**interp(log10(mvals),log10(m1),log10(d_e1))
dvals2 = 10**interp(log10(mvals),log10(m2),log10(d_e2))
dvals3 = 10**interp(log10(mvals),log10(m3),log10(d_e3))
dvals2[mvals>amax(m2)] = inf
dvals1[mvals>amax(m1)] = inf
dvals = minimum(dvals1,dvals2)
dvals = minimum(dvals,dvals3)
plt.plot(mvals,dvals,lw=4)

DAT = column_stack((mvals,dvals))
savetxt('limit_data/ScalarPhoton/FifthForce.txt',DAT,
        header='Composition-independent fifth force combined constraints \n Union of ScalarBaryon/FifthForce_CI.txt and ScalarBaryon_Mars.txt \n mass [eV] \t d_e')
```
