```python
fig,ax = FigSetup(Shape='Square',ylab='$|g_{a\gamma\gamma}|$ [GeV$^{-1}$]',mathpazo=True,
                  m_min=1e6,m_max=2e12,g_min=1e-8,g_max=1.0e-1,xtick_rotation=0)

def FASER(ax,text_shift=[1,1],col='k',text_col='k',fs=16.5,zorder=10,text_on=True,lw=1.5,rotation=-75,ha='center',edgealpha=1,path_effects=[]):
    dat = loadtxt("limit_data/AxionPhoton/Projections/FASER_LLP.txt")
    plt.plot(dat[:,0],dat[:,1],'--',lw=lw,color=col,alpha=edgealpha,zorder=zorder)

    if text_on:
        plt.text(text_shift[0]*1.2e8,text_shift[1]*2e-4,r'{\bf FASER (LLP)}',fontsize=fs,color=text_col,rotation=rotation,ha='center',va='top',clip_on=True,path_effects=path_effects)
    return

def CONUS(ax,text_shift=[1,1],col='k',text_col='k',fs=16.5,zorder=10,text_on=True,lw=1.5,rotation=-20,ha='center',edgealpha=1,path_effects=[]):
    dat = loadtxt("limit_data/AxionPhoton/Projections/CONUS.txt")
    plt.plot(dat[:,0],dat[:,1],'--',lw=lw,color=col,alpha=edgealpha,zorder=zorder)

    if text_on:
        plt.text(text_shift[0]*2.2e6,text_shift[1]*1.3e-6,r'{\bf CONUS ($\nu$)}',fontsize=fs,color=text_col,rotation=rotation,ha='center',va='top',clip_on=True,path_effects=path_effects)
    return

def DUNE(ax,text_shift=[1,1],col='k',text_col='k',fs=16.5,zorder=10,text_on=True,lw=1.5,rotation=-65,ha='center',edgealpha=1,path_effects=[]):
    dat = loadtxt("limit_data/AxionPhoton/Projections/DUNE.txt")
    plt.plot(dat[:,0],dat[:,1],'--',lw=lw,color=col,alpha=edgealpha,zorder=zorder)

    if text_on:
        plt.text(text_shift[0]*1.1e9,text_shift[1]*1e-6,r'{\bf DUNE-like} (decay)',fontsize=fs,color=text_col,rotation=rotation,ha='center',va='top',clip_on=True,path_effects=path_effects)
    return

DUNE(ax)
CONUS(ax)
FASER(ax)

# to add:
# CLIC
# FCC
# SHIPS
# DarkQuest
# Belle II
# LHCb?

AxionPhoton.SNe_decay(ax,rotation=0,fs=30,text_pos=[0.7e7,3e-8])
AxionPhoton.BeamDump(ax,rotation=-45,fs=33,text_shift=[0.9,0.3])
AxionPhoton.BaBar(ax,fs=25,text_shift=[3.5,0.4])
AxionPhoton.CMS_PbPb(ax,fs=30,text_shift=[1,0.8])
AxionPhoton.ATLAS_PbPb(ax,fs=30,text_shift=[1,1.2])
AxionPhoton.LHC_pp(ax,fs=28,text_col='w',text_shift=[1.7,10],path_effects=line_background(1.5,'k'))
AxionPhoton.BelleII(ax,fs=30)
AxionPhoton.PrimEx(ax,fs=20,text_shift=[0.85,0.9],rotation=90)
AxionPhoton.LEP(ax,fs=35,text_shift=[1,0.2])
AxionPhoton.BESIII(ax,fs=31,text_shift=[1.3,1.4])
AxionPhoton.OPAL(ax,fs=24,text_shift=[0.021,3.5],text_col='w',path_effects=line_background(1.5,'k'))


MySaveFig(fig,'AxionPhoton_ColliderBounds_with_Projections')
```
